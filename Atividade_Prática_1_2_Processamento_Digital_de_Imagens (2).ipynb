```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from urllib.request import urlretrieve

plt.rcParams["figure.figsize"] = (8, 6)
plt.rcParams["axes.grid"] = False
```

## 0.1 Download das imagens diretamente do GitHub

> **Importante:** antes de executar este notebook no Colab, envie para o seu repositório a pasta `imagens/` com os arquivos `img01.jpeg` até `img10.jpeg`.

```python
GITHUB_USER = "joaofelipeunemat"
GITHUB_REPO = "Processamento-de-imagem"
BRANCH = "main"
BASE_URL = f"https://raw.githubusercontent.com/{GITHUB_USER}/{GITHUB_REPO}/{BRANCH}"

arquivos = [f"img{i:02d}.jpeg" for i in range(1, 11)]

os.makedirs("imagens", exist_ok=True)

for nome in arquivos:
    url = f"{BASE_URL}/{nome}"
    destino = os.path.join("imagens", nome)
    urlretrieve(url, destino)

print("Download concluído com sucesso.")
print(os.listdir("imagens"))
```

## 0.2 Funções auxiliares

```python
def mostrar(img, titulo="", cmap=None):
    plt.figure(figsize=(7, 5))
    if len(img.shape) == 3 and cmap is None:
        plt.imshow(img)
    else:
        plt.imshow(img, cmap=cmap)
    plt.title(titulo)
    plt.axis("off")
    plt.show()

def carregar_colorida(caminho):
    img_bgr = cv2.imread(caminho)
    if img_bgr is None:
        raise FileNotFoundError(f"Imagem não encontrada: {caminho}")
    return img_bgr

def bgr_para_rgb(img_bgr):
    return cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

## Exercício 1 - Carregar uma imagem colorida própria e exibi-la com correção de canais (BGR → RGB)

```python
img_bgr = carregar_colorida("imagens/img01.jpeg")
img_rgb = bgr_para_rgb(img_bgr)

mostrar(img_rgb, "Exercício 1 - Imagem original com correção de canais (RGB)")
```

## Exercício 2 - Imprimir os atributos `shape`, `size` e `dtype` da imagem

```python
print("Shape:", img_bgr.shape)
print("Size:", img_bgr.size)
print("Dtype:", img_bgr.dtype)
```

## Exercício 3 - Salvar uma cópia da imagem original em um formato de compressão diferente

```python
cv2.imwrite("img01_convertida.jpg", img_bgr)
print("Arquivo salvo com sucesso como img01_convertida.jpg")
```

## Exercício 4 - Acessar as coordenadas centrais da imagem e exibir os valores R, G e B do pixel

```python
altura, largura, _ = img_rgb.shape
centro_y = altura // 2
centro_x = largura // 2

pixel_rgb = img_rgb[centro_y, centro_x]
print(f"Coordenada central: ({centro_x}, {centro_y})")
print(f"Valores RGB do pixel central: R={pixel_rgb[0]}, G={pixel_rgb[1]}, B={pixel_rgb[2]}")
```

## Exercício 5 - Modificar uma região de 100 × 100 no canto superior para uma cor sólida e exibir o resultado

```python
img_ex5 = img_rgb.copy()
img_ex5[0:100, 0:100] = [255, 0, 0]  # vermelho em RGB

mostrar(img_ex5, "Exercício 5 - Região 100x100 alterada para vermelho")
```

## Exercício 6 - Realizar um recorte (ROI) focando em um detalhe da foto e exibir apenas o recorte

```python
h, w, _ = img_rgb.shape
roi = img_rgb[h//4:h//4 + h//3, w//4:w//4 + w//3]

mostrar(roi, "Exercício 6 - Recorte (ROI)")
print("Shape da ROI:", roi.shape)
```

## Exercício 7 - Separar os canais B, G e R da imagem utilizando funções do OpenCV

```python
b, g, r = cv2.split(img_bgr)

print("Canal B:", b.shape)
print("Canal G:", g.shape)
print("Canal R:", r.shape)
```

## Exercício 8 - Exibir o canal Verde (G) isoladamente como imagem em tons de cinza

```python
mostrar(g, "Exercício 8 - Canal Verde isolado", cmap="gray")
```

## Exercício 9 - Realizar o merge dos canais trocando a posição do R com o B

```python
img_trocada_bgr = cv2.merge([r, g, b])
img_trocada_rgb = cv2.cvtColor(img_trocada_bgr, cv2.COLOR_BGR2RGB)

mostrar(img_trocada_rgb, "Exercício 9 - Imagem com canais R e B trocados")
```

## Exercício 10 - Converter a imagem original para tons de cinzento via software

```python
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

mostrar(img_gray, "Exercício 10 - Imagem em tons de cinza", cmap="gray")
```

## Exercício 11 - Converter a imagem para o espaço HSV e extrair apenas o canal de Saturação (S)

```python
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
h_canal, s_canal, v_canal = cv2.split(img_hsv)

mostrar(s_canal, "Exercício 11 - Canal de Saturação (S)", cmap="gray")
```

## Exercício 12 - Aumentar o brilho da imagem somando 50 unidades aos pixels com `cv2.add`

```python
aumento = np.full(img_bgr.shape, 50, dtype=np.uint8)
img_brilho_bgr = cv2.add(img_bgr, aumento)
img_brilho_rgb = cv2.cvtColor(img_brilho_bgr, cv2.COLOR_BGR2RGB)

mostrar(img_brilho_rgb, "Exercício 12 - Imagem com brilho aumentado")
```

## Exercício 13 - Redimensionar duas fotos para o mesmo tamanho e fazer a mesclagem com `addWeighted`

```python
img2_bgr = carregar_colorida("imagens/img02.jpeg")

altura1, largura1 = img_bgr.shape[:2]
img2_redim_bgr = cv2.resize(img2_bgr, (largura1, altura1))

mistura_bgr = cv2.addWeighted(img_bgr, 0.5, img2_redim_bgr, 0.5, 0)
mistura_rgb = cv2.cvtColor(mistura_bgr, cv2.COLOR_BGR2RGB)

mostrar(mistura_rgb, "Exercício 13 - Mesclagem de duas imagens (0.5 / 0.5)")
```

## Exercício 14 - Implementar o negativo fotográfico da imagem em tons de cinzento

```python
negativo = 255 - img_gray

mostrar(negativo, "Exercício 14 - Negativo fotográfico", cmap="gray")
```

## Exercício 15 - Criar uma máscara binária circular e usar para isolar uma parte da foto principal

```python
h, w = img_bgr.shape[:2]
mask = np.zeros((h, w), dtype=np.uint8)

centro = (w // 2, h // 2)
raio = min(h, w) // 4

cv2.circle(mask, centro, raio, 255, -1)

resultado_mascara = cv2.bitwise_and(img_rgb, img_rgb, mask=mask)

mostrar(mask, "Exercício 15 - Máscara circular", cmap="gray")
mostrar(resultado_mascara, "Exercício 15 - Região isolada com máscara")
```

## Exercício 16 - Reduzir a imagem original para 30% do tamanho inicial e exibir o novo `shape`

```python
img_30 = cv2.resize(img_bgr, (0, 0), fx=0.3, fy=0.3)
img_30_rgb = cv2.cvtColor(img_30, cv2.COLOR_BGR2RGB)

print("Novo shape:", img_30.shape)
mostrar(img_30_rgb, "Exercício 16 - Imagem reduzida para 30%")
```

## Exercício 17 - Calcular e imprimir os valores de intensidade do histograma da imagem

```python
hist_gray = cv2.calcHist([img_gray], [0], None, [256], [0, 256])

print("Valores do histograma em tons de cinza:")
for i, valor in enumerate(hist_gray.flatten()):
    print(f"Intensidade {i}: {int(valor)}")
```

## Exercício 18 - Plotar o histograma dos canais de cores separadamente no mesmo gráfico

```python
cores = ("b", "g", "r")
plt.figure(figsize=(10, 5))

for i, cor in enumerate(cores):
    hist = cv2.calcHist([img_bgr], [i], None, [256], [0, 256])
    plt.plot(hist, color=cor, label=f"Canal {cor.upper()}")

plt.title("Exercício 18 - Histograma dos canais B, G e R")
plt.xlabel("Intensidade")
plt.ylabel("Quantidade de pixels")
plt.legend()
plt.show()
```

## Exercício 19 - Aplicar Equalização de Histograma para melhorar o contraste de uma imagem escura

```python
img_escura_bgr = carregar_colorida("imagens/img03.jpeg")
img_escura_gray = cv2.cvtColor(img_escura_bgr, cv2.COLOR_BGR2GRAY)
img_equalizada = cv2.equalizeHist(img_escura_gray)

mostrar(img_escura_gray, "Exercício 19 - Imagem escura original", cmap="gray")
mostrar(img_equalizada, "Exercício 19 - Imagem com equalização de histograma", cmap="gray")
```

## Exercício 20 - Aplicar um limiar (Threshold) binário simples para separar o objeto do fundo

```python
img_thresh_bgr = carregar_colorida("imagens/img04.jpeg")
img_thresh_gray = cv2.cvtColor(img_thresh_bgr, cv2.COLOR_BGR2GRAY)

_, th_binaria = cv2.threshold(img_thresh_gray, 127, 255, cv2.THRESH_BINARY)

mostrar(img_thresh_gray, "Exercício 20 - Imagem em tons de cinza", cmap="gray")
mostrar(th_binaria, "Exercício 20 - Threshold binário", cmap="gray")
```
